```python
import os
import json
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torchvision.transforms.functional as F
```

```python
from tumor_dataset import TumorDataset

tumor_dataset = TumorDataset('BraTS2020_TrainingData/MICCAI_BraTS2020_TrainingData')

figure, axis = plt.subplots(3, 5, figsize=(12, 8)) 
for i in range(3):
    t1 = tumor_dataset[i][0][:, :, 65].detach()
    t1ce = tumor_dataset[i][1][:, :, 65].detach()
    t2 = tumor_dataset[i][2][:, :, 65].detach()
    flair = tumor_dataset[i][3][:, :, 65].detach()
    seg = tumor_dataset[i][4][:, :, 65].detach()
    t1 = F.to_pil_image(t1)
    t1ce = F.to_pil_image(t1ce)
    t2 = F.to_pil_image(t2)
    flair = F.to_pil_image(flair)
    seg = F.to_pil_image(seg)
    axis[i, 0].imshow(np.asarray(t1))
    axis[i, 1].imshow(np.asarray(t1ce))
    axis[i, 2].imshow(np.asarray(t2))
    axis[i, 3].imshow(np.asarray(flair))
    axis[i, 4].imshow(np.asarray(seg))
    axis[i, 0].set(xticklabels=[], yticklabels=[], xticks=[], yticks=[])
    axis[i, 1].set(xticklabels=[], yticklabels=[], xticks=[], yticks=[])
    axis[i, 2].set(xticklabels=[], yticklabels=[], xticks=[], yticks=[])
    axis[i, 3].set(xticklabels=[], yticklabels=[], xticks=[], yticks=[])
    axis[i, 4].set(xticklabels=[], yticklabels=[], xticks=[], yticks=[])
    axis[i, 0].set_ylabel(f'sample {i} \nslice 65')
axis[0, 0].set_title('t1 scan')
axis[0, 1].set_title('t1ce scan')
axis[0, 2].set_title('t2 scan')
axis[0, 3].set_title('flair scan')
axis[0, 4].set_title('target')
figure.tight_layout()
```
